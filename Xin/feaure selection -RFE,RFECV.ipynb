# 1. RFECV method 

We can use RFECV method to get the numebers of the best feature seletion rusult and also which these features are.

```python
# Use RFECV to know the best numbers of features and which are the best features
from sklearn.datasets import make_friedman1
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LinearRegression

X = train_minmax_.values
Y = train_minmax["logerror"]
# Create the RFE object and compute a cross-validated score.
model = LinearRegression ()
selector = RFECV(model, step=1, cv=None)
selector = selector.fit(X, Y)
selector.support_ 
```

```python
print("Optimal number of features : %d" % selector.n_features_)
```

```python
selector.ranking_
```

```python
featureindex = np.where(selector.ranking_==1)
features= train_minmax_.columns
```

```python
print('The best features subset is:')
print(features[featureindex])
```

As we can see here, the best features by RFECV method are the subset of 

['finishedsquarefeet15', 'fips', 'rawcensustractandblock','taxvaluedollarcnt', 'taxamount', 'taxdelinquencyyear']


# 2.RFE method 

We can also ues RFE method to rank the importance of the feature. As we know, the number of the best feature is 6.We can choose the first 6 features of the rank then we get the best subset of features.

```python
#feature selection by Wrappers method (RFE)

from pandas import read_csv
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression 

features= train_minmax_.columns
X = train_minmax_.values
Y = train_minmax["logerror"]
# feature extraction
model = LinearRegression ()

rfe = RFE(model, n_features_to_select=6)  
rfe.fit(X,Y)
print ("Features sorted by their rank:")  
print (sorted(zip(map(lambda x: round(x, 4), rfe.ranking_), features))) 
```

As we can see here, the fisrt 6 features of the rank is the same as the RFECV's result.By conclusion, When we use RFE oe RFECV method, the best subset of feature is 

['finishedsquarefeet15', 'fips', 'rawcensustractandblock','taxvaluedollarcnt', 'taxamount', 'taxdelinquencyyear']
