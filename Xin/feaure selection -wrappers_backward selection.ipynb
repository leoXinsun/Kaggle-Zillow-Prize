```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
train_minmax = pd.read_csv('train_minmax.csv')
```

```python
train_minmax
```

```python
train_minmax_ = train_minmax.drop(['logerror','logerror_labels','Unnamed: 0','Unnamed: 0.1'],1)
train_minmax_
```

```python
from sklearn.cross_validation import cross_val_score
from sklearn.linear_model import LinearRegression
model = LinearRegression ()
X = train_minmax_.values
Y = train_minmax["logerror"]
F= train_minmax_.columns
f = len(F)

score = (cross_val_score(model, X, Y, cv=None)) ** 2
score0 = score.mean()
score = score0
```

# Wrappers - backward selection
In backward selection, one starts with the set of all variables and progressively eliminates the least promising ones.

```python
for i in range(f):
    # find which feature make the value of the evaluation function is optimal when remove this feature
    for j in range(len(F)):  
        train_minmax_1 = train_minmax_.drop([F[j]],1, inplace = False)
        X1 = train_minmax_1.values 
        score1_ = (cross_val_score(model, X1, Y, cv=None)) ** 2
        score1 = score1_.mean()
        if score1 < score:
            score = score1
            k = j
    # remove this feature
    if score0 != score:         
        train_minmax_ = train_minmax_.drop([F[k]],1)
        F = F.drop(F[k],1)
        score0 = score
    # when we remove anyone feature, the value of the evaluation function cannot be better , we get the best feature
    else:   
        break
print(F)              
```
