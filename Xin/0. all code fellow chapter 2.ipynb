# 1. Bulid DataFrame

```python
import pandas as pd
import numpy as np

train_2016 = pd.read_csv('train_2016_v2.csv')
train_2017 = pd.read_csv('train_2017.csv')
properties_2016 = pd.read_csv('properties_2016.csv')
properties_2017 = pd.read_csv('properties_2017.csv')
dataset_2016 = pd.merge(train_2016, properties_2016, on='parcelid', how='left')
dataset_2017 = pd.merge(train_2017, properties_2017, on='parcelid', how='left')
dataset = pd.merge(dataset_2016, dataset_2017, how='outer')
```

# 6. Data cleaning

```python
dataset.head()
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import Imputer

imputer = Imputer(strategy = "median")
dataset_num = dataset.drop(["transactiondate", "hashottuborspa",
                            "propertycountylandusecode","propertyzoningdesc",
                           "fireplaceflag","taxdelinquencyflag"], axis=1)
imputer.fit(dataset_num)
imputer.statistics_
```

```python
dataset_num.median().values
```

```python
X = imputer.transform(dataset_num)
dataset_tr = pd.DataFrame(X, columns = dataset_num.columns)
```

# 7. Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler 

train_mm = MinMaxScaler().fit_transform(dataset_tr.values)

train_minmax=pd.DataFrame(train_mm,columns=dataset_tr.columns)

train_minmax
```

# 8. Training and Evaluating

```python
# Linear Regression
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
train_labels = train_minmax["logerror"].copy()
lin_reg.fit(train_minmax, train_labels)
```

```python
from sklearn.metrics import mean_squared_error
train_predictions = lin_reg.predict(train_minmax)
lin_mse = mean_squared_error(train_labels, train_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
# DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(train_minmax,train_labels)
```

```python
train_predictions = tree_reg.predict(train_minmax)
tree_mse = mean_squared_error(train_labels, train_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

# 9. Better Evaluation Using Cross-Validation

```python
# DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg,train_minmax,train_labels,
                        scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
tree_rmse_scores 
```

```python
def display_scores(scores):
    print("scores:",scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
display_scores(tree_rmse_scores)
```

```python
#for the Linear Regression model
lin_scores = cross_val_score(lin_reg,train_minmax,train_labels,
                        scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

Form above, we can know the Linear Regression model preforms better than DecisionTreeRegressor.

Try tha last model: RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(train_minmax, train_labels)
```

```python
train_predictions = forest_reg.predict(train_minmax)
forest_mse = mean_squared_error(train_labels, train_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
forest_scores = cross_val_score(forest_reg,train_minmax,train_labels,
                        scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```
